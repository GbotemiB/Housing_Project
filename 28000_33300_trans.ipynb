#next thing to do is to import transformer library and pipeline to use huggingface nlp models.

```python
import transformers
from transformers import pipeline
```

```python
#using zero shot classification 

classifier = pipeline("zero-shot-classification")
```

#The description function is take as input the translated text of a row, score it using a classifier from huggingface and determine if it is a rental or not rental statement.

```python
def description(text):
  results = classifier(text, candidate_labels = ['not_rent', 'rentals'])
  return results['scores'][0]
```

```python
description(data['transDescription'][12])
```

```python
data['score'] = 0.0
```

#The for loop on the next line is used to get the description score and append the result to a file in the drive. This was done in batches due to computation cost. it takes more than 12 hours to get a score for up to 4000 rows. So this process was done in batches in multiple notebooks to save time which still took up to 3 days with just 99,000 rows.

```python
for i in range(28000, 33300):
  data['score'][i] = description(data['transDescription'][i])
  print(i, data['score'][i])

  data.loc[[i]].to_csv('drive/MyDrive/28000_trans_new.csv', mode='a', index=False, header=False)
```
