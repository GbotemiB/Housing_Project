```python
import transformers
from transformers import pipeline
```

```python
#using zero shot classification 

classifier = pipeline("zero-shot-classification")
```

```python
def description(text):
  results = classifier(text, candidate_labels = ['not_rent', 'rentals'])
  return results['scores'][0]
```

```python
description(data['transDescription'][12])
```

```python
data['score'] = 0.0
```

```python
for i in range(3545, 7000):
  data['score'][i] = description(data['transDescription'][i])
  print(i, data['score'][i])

  data.loc[[i]].to_csv('drive/MyDrive/7000_trans_new.csv', mode='a', index=False, header=False)
```
