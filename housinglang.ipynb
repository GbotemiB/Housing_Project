```python
import re

CLEANR = re.compile('<.*?>')

def cleanhtml(rawhtml):
    cleantext = re.sub(CLEANR, '', rawhtml)
    cleantext = re.sub(r'http\S+', '', cleantext)
    return cleantext
    
```
